```python
from typing import Annotated
```

```python
class Creato(TypedDict):
  x : Annotated[int, "you extract how many people are there"]
  n : Annotated[str, " you extract their body condition"]
  v : Annotated[float, "you tell their avg height"]
```

```python
modelk = ChatGoogleGenerativeAI(model = "models/gemini-2.5-flash" )
xyz = modelk.with_structured_output(Creato)
resu = xyz.invoke("there are 5 people, i am healthy and thier avg height is 7.9")
resu
```
